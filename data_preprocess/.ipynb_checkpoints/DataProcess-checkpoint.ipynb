```python
import collections
import json

foodClass = {"cheesecake", "mussels", "waffles", "pizza",  "steak", "tacos", "donuts", "sushi"}


photoComment = dict()
photoDataset = collections.defaultdict(list)
photobusiness = collections.defaultdict(list)

with open("./yelp_photos/photos.json") as json_file:
    result = []
    for line in json_file.readlines():
        data = json.loads(line)
        if  data["caption"]:
            photoComment[data["photo_id"]]= data["caption"]
            
            
            photoId = data["photo_id"]
            caption = data["caption"]
            businessid = data["business_id"]
        
            caption = caption.replace("\n", " ")
            caption = caption.replace("\r", " ")
            caption = caption.replace("\r\n", " ")
            
            for character in ",.!:@&-/()+?":
                caption = caption.replace(character, " ")  
                
            for food in foodClass:
                if businessid in photobusiness[food]:
                    photoDataset[food].append(photoId)

                
                
            words = caption.split()
            words = [caption.lower() for word in words]

            if "fried" in words or "rice" in words:
                photoDataset["fried_rice"].append(photoId)
                continue
            for word in words:
                if len(word)<2:
                    continue
                wordminuss = word[:len(word)-1]
                wordpluss = word + "s"
                if word in foodClass:
                    photoDataset[word.lower()].append(photoId)
                    photobusiness[word.lower()].append(photoId)
                elif wordpluss in foodClass:
                    photoDataset[wordpluss.lower()].append(photoId)
                    photobusiness[wordpluss.lower()].append(photoId)

                elif wordminuss in foodClass:
                    photoDataset[wordminuss.lower()].append(photoId)
                    photobusiness[wordminuss.lower()].append(photoId)

            
```
